```python
# Dependencies and Setup
# SQL Alchemy
from sqlalchemy import create_engine

# Pandas
import pandas as pd

# Matplotlib
import matplotlib.pyplot as plt

# NumPy
import numpy as np
```

```python
# Create Engine and Pass in Postgres Connection
# Setup to Connect to Database 
engine = create_engine('postgres://postgres:password@localhost:5432/employee_db')
conn = engine.connect()
```

```python
# Query all records in join salaries and employee table
salaries_data = pd.read_sql("SELECT employees.emp_no, employees.emp_title_id, salaries.salary FROM employees JOIN salaries ON employees.emp_no = salaries.emp_no;", conn)
salaries_data.head()
```

```python
# Query all records in the titles talbel
titles_data = pd.read_sql("SELECT title_id  AS emp_title_id, title FROM titles", conn)
titles_data.head()
```

```python
# Merge Columns / Combine Data
combined_data = pd.merge(salaries_data, titles_data, on="emp_title_id", how="inner")
combined_data.head()
```

```python
# Groupby Title
grouped_df = combined_data.groupby("title").mean()
grouped_df
```

```python
# Drop emp_no from grouped_df to Show Only title & salary
revised_grouped_df = grouped_df.drop(columns = "emp_no")
revised_grouped_df= revised_grouped_df.reset_index()
revised_grouped_df=revised_grouped_df.sort_values('salary')
revised_grouped_df
```

```python
# Set x_axis, y_axis & Tick Locations
x_axis = revised_grouped_df["title"]
ticks = np.arange(len(x_axis))
y_axis = revised_grouped_df["salary"]
 
# Create Bar Chart Based on Data
plt.bar(x_axis, y_axis, align="center", color=["pink", "red", "orange", "yellow", "green", "blue", "indigo"])

# Create Ticks for Bar Chart's x_axis
plt.xticks(ticks, x_axis, rotation="vertical")

# Plot properties
plt.ylabel("Salaries ($)")
plt.xlabel("Employee Titles")
plt.title("Average Employee Salary by Title")

# Save Figure
plt.savefig("./Images/bar plot.png")

# Show plot
plt.show()
```
